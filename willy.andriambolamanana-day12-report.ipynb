### Exercise 11:

### Exercise 17:
Create a plot with multiple subplots in a 2x2 grid. Plot `y1 = x`, `y2 = x^2`, `y3 = x^3`, and `y4 = x^4`.

```python
# Exercise 17
```

---

````python
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

x = np.linspace(0, 10, 100)
y1 = x
y2 = x**2
y3 = x**3
y4 = x**4

# Plot y1 = x
axs[0, 0].plot(x, y1, 'b-')
axs[0, 0].set_title('y = x')

# Plot y2 = x^2
axs[0, 1].plot(x, y2, 'r-')
axs[0, 1].set_title('y = x^2')

# Plot y3 = x^3
axs[1, 0].plot(x, y3, 'g-')
axs[1, 0].set_title('y = x^3')

# Plot y4 = x^4
axs[1, 1].plot(x, y4, 'm-')
axs[1, 1].set_title('y = x^4')

# Add a main title and adjust layout
fig.suptitle('Multiple Subplots in a 2x2 Grid')
plt.tight_layout()
plt.subplots_adjust(top=0.9)
plt.show()
````

### Exercise 21:
Plot a contour plot for the function `z = sin(x) + cos(y)` over a grid of `x` and `y` values ranging from -5 to 5.

```python
# Exercise 21
```

---

````python
# Generate x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Compute z values
Z = np.sin(X) + np.cos(Y)

# Create the contour plot
plt.figure(figsize=(8, 6))
contour = plt.contour(X, Y, Z, levels=20, cmap='viridis')
plt.colorbar(contour)  # Show color scale
plt.title('Contour Plot of z = sin(x) + cos(y)')
plt.xlabel('x values')
plt.ylabel('y values')
plt.show()
````

### Exercise 26:
Create a 3D plot of the function `z = x^2 + y^2` over a grid of `x` and `y` values.

```python
# Exercise 26
```

---

````python
# Create a grid of x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis')

# Add labels
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D plot of z = x^2 + y^2')

# Show the plot
plt.show()
````

### Exercise 27:
Create a radar chart with 5 variables and random values between 0 and 1.

```python
# Exercise 27
```

---

````python
# Number of variables
num_vars = 5

# Create a 1D array of angles for each variable
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# Make the plot close to a circular shape
values = np.random.rand(num_vars).tolist()
values += values[:1]  # repeat the first value to close the circle
angles += angles[:1]  # repeat the first angle to close the circle

# Create the radar chart
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, values, color='blue', alpha=0.25)
ax.plot(angles, values, color='blue', linewidth=2)

# Add labels for each variable
labels = [f'Var {i+1}' for i in range(num_vars)]
plt.xticks(angles[:-1], labels)

# Show the plot
plt.title('Radar Chart with Random Values')
plt.show()
````

### Exercise 28:
Create a violin plot to show the distribution of a dataset generated with 1000 random numbers from a normal distribution.

```python
# Exercise 28
```

---

````python

# Generate a dataset of 1000 random numbers from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Create a figure
plt.figure(figsize=(8, 6))

# Create the violin plot
parts = plt.violinplot(data, showmeans=False, showmedians=True)

# Customize the plot
for pc in parts['bodies']:
    pc.set_facecolor('skyblue')
    pc.set_edgecolor('black')
    pc.set_alpha(0.7)

# Add labels and title
plt.title('Violin Plot of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Density')

# Show the plot
plt.show()
````
